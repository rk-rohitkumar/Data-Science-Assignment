```python
# Task 2: Mark the success/failed launches for each site on the map
import folium
from folium.plugins import MarkerCluster

# Center coordinate for initializing map (NASA Johnson Space Center example)
map_center = [29.559684888503615, -95.0830971930759]

# Initialize a fresh new map for Task 2 visualization
launch_outcome_map = folium.Map(location=map_center, zoom_start=5)

# Create a MarkerCluster to group close launch points
marker_cluster = MarkerCluster().add_to(launch_outcome_map)

# Iterate over each launch record to add colored markers based on success or failure
for index, row in spacex_df.iterrows():
    lat = row['Lat']
    long = row['Long']
    outcome_class = row['class']  # 1 = success, 0 = failure
    launch_site = row['Launch Site']

    # Set marker color for success (green) and failure (red)
    marker_color = 'green' if outcome_class == 1 else 'red'
    popup_text = f"{launch_site}<br>Status: {'Success' if outcome_class == 1 else 'Failure'}"

    # Add marker to the cluster on the map
    folium.Marker(
        location=[lat, long],
        popup=folium.Popup(popup_text, max_width=300),
        icon=folium.Icon(color=marker_color)
    ).add_to(marker_cluster)

# Display the map with launch success/failure markers
launch_outcome_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from folium.features import DivIcon
import math

# Coordinates for launch site and closest coastline point
launch_site_coord = (28.5721, -80.648)  
coastline_coord = (28.56367, -80.57163)  # closest coastline point

# Haversine formula to calculate distance between two lat/lon points in kilometers
def haversine(coord1, coord2):
    R = 6371  # Earth radius in km
    lat1, lon1 = math.radians(coord1[0]), math.radians(coord1[1])
    lat2, lon2 = math.radians(coord2[0]), math.radians(coord2[1])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
    c = 2 * math.asin(math.sqrt(a))
    return R * c

distance = haversine(launch_site_coord, coastline_coord)

# Add a marker for the coastline point with distance displayed
distance_marker = folium.Marker(
    location=coastline_coord,
    icon=DivIcon(
        icon_size=(150,36),
        icon_anchor=(0,0),
        html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' 
             % "{:10.2f} KM".format(distance),
    )
)

# Add marker to your map object
site_map.add_child(distance_marker)
```

```python
poi_locations =  {
    'CCAFS SLC-40': {
        'highway': (28.40416, -80.6296),
        
        'coastline': (28.56425, -80.56813),
    },
    'CCAFS LC-40': {
        'highway': (28.40416, -80.6296),
        
        'coastline': (28.56425, -80.56813),
    }
}

import math

def haversine(coord1, coord2):
    R = 6371  # Earth radius in km
    lat1, lon1 = math.radians(coord1[0]), math.radians(coord1[1])
    lat2, lon2 = math.radians(coord2[0]), math.radians(coord2[1])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = math.sin(dlat / 2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2)**2
    c = 2 * math.asin(math.sqrt(a))
    return R * c
from folium.features import DivIcon
import folium

for _, row in spacex_df.iterrows():
    site_name = row['Launch Site']
    site_coord = (row['Lat'], row['Long'])
    
    # Add launch site marker
    folium.Marker(location=site_coord, popup=site_name).add_to(site_map)
    
    # For each POI type near this launch site
    if site_name in poi_locations:
        for poi_type, poi_coord in poi_locations[site_name].items():
            # Add POI marker
            folium.Marker(location=poi_coord, popup=poi_type).add_to(site_map)
            
            # Draw lines between site and POI
            line = folium.PolyLine(locations=[site_coord, poi_coord], color='blue', weight=1)
            site_map.add_child(line)
            
            # Calculate and display distance
            dist = haversine(site_coord, poi_coord)
            mid_lat = (site_coord[0] + poi_coord[0]) / 2
            mid_lon = (site_coord[1] + poi_coord[1]) / 2
            
            folium.Marker(
                location=[mid_lat, mid_lon],
                icon=DivIcon(icon_size=(150,36), icon_anchor=(0,0),
                             html=f'<div style="font-size: 12px; color:#d35400;"><b>{dist:.2f} km</b></div>')
            ).add_to(site_map)
site_map
```
